```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, filtfilt

# Assuming 'data' is the variable you want to plot against time
# Generate a time array based on the length of trial1_time_all_channel_1
time_data=(info_data['TimeInSec'])
time=(time_data[0][0][0])

# Define a function for a Butterworth low-pass filter
def butter_lowpass_filter(data, cutoff_freq, sample_rate, order=4):
    nyquist = 0.5 * sample_rate
    normal_cutoff = cutoff_freq / nyquist
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    y = filtfilt(b, a, data)
    return y

# Define filter parameters
cutoff_frequency = 10  # Adjust this based on your data
sample_rate = 1000  # Adjust this based on your data

# Apply the low-pass filter to denoise the signal
denoised_signal = butter_lowpass_filter(trial1_time_all_channel_1, cutoff_frequency, sample_rate)
plt.figure(figsize=(10, 6))

# Plot the original and denoised signals
plt.plot(time, trial1_time_all_channel_1, label='Original Signal')
plt.plot(time, denoised_signal, label='Denoised Signal', linewidth=2)
plt.title('Original and Denoised Signals')
plt.xlabel('Time (samples)')
plt.ylabel('Data Values')
plt.legend()
plt.show()
```
